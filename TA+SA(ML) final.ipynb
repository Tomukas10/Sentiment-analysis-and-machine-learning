```python
atr = AverageTrueRange(high=data["High"], low=data["Low"], close=data["Close"], window=14)
data["ATR_14"] = atr.average_true_range()
```

```python
#split the train and test data and replace any nan's to avoid reducing any nan train set rows (it  hurts accuracy)
data = data.fillna(0.0)
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size].copy()
test_data = data.iloc[train_size:].copy()
```

```python
#set targets
train_data["Target"] = (train_data["Close"].shift(-1) > train_data["Close"]).astype(int)

# drop last row in train set to avoid erroneus data
train_data = train_data[:-1]

# Features
feature_cols = ["EMA_14", "RSI_14", "MACD", "MACD_Signal", "MACD_Hist", "ATR_14"]
x_train = train_data[feature_cols]
y_train = train_data["Target"]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

print(data)
```

```python
# set target for test data 
test_data["Target"] = (test_data["Close"].shift(-1) > test_data["Close"]).astype(int)
test_data = test_data[:-1]  # drop last row

X_test = test_data[feature_cols]
y_test = test_data["Target"]

# Predict
preds = model.predict(X_test)

# print accuracy
from sklearn.metrics import accuracy_score
print("Test Accuracy:", accuracy_score(y_test, preds))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, preds))
print("Confusion Matrix:\n", confusion_matrix(y_test, preds))
```

```python
#combined TA + SA model
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

#all n, y convert to 0 and 1's
def yn_num(val):
    v = str(val).strip().lower()
    if v == 'y':
        return 1.0
    elif v == 'n':
        return 0.0
    else:
        return np.nan  

data["Results"] = data["Results"].apply(yn_num)
data["Results_plus"] = data["Results(+)"].apply(yn_num)
data["Sentiment"] = pd.to_numeric(data["Sentiment"], errors="coerce")

#Create feature set
feature_cols_all = ["EMA_14", "RSI_14", "MACD", "MACD_Signal", "MACD_Hist", "ATR_14", "Sentiment", "Results", "Results_plus"]

#Fill any NaNs with neutral sentiment 
data_set = data.copy()
data_set["Sentiment"] = data_set["Sentiment"].fillna(0.0)
data_set["Results"] = data_set["Results"].fillna(0.5)
data_set["Results_plus"] = data_set["Results_plus"].fillna(0.5)

data_set[["EMA_14","RSI_14","MACD","MACD_Signal","MACD_Hist","ATR_14"]] = data_set[["EMA_14","RSI_14","MACD","MACD_Signal","MACD_Hist","ATR_14"]].fillna(0.0)

# Create target, 1 = next day Close higher, 0 = lower
data_set["Target"] = (data_set["Close"].shift(-1) > data_set["Close"]).astype(int)
data_set = data_set[:-1]

# Train test split 80/20 (Practically the semptemeber month is the prediction set)
train_size = int(len(data_set) * 0.8)
train_data = data_set.iloc[:train_size]
test_data = data_set.iloc[train_size:]

#train
x_train_all = train_data[feature_cols_all]
y_train_all = train_data["Target"]

#test
X_test_all = test_data[feature_cols_all]
y_test_all = test_data["Target"]

# Train model
model_all = LogisticRegression(max_iter=1000)
model_all.fit(x_train_all, y_train_all)
preds_all = model_all.predict(X_test_all)
```

```python
# Results and plot
ta_acc = accuracy_score(y_test, preds)
sa_ta_acc = accuracy_score(y_test_all, preds_all)

#accuracy comaparison
print(f"TA Accuracy:       {ta_acc*100:.1f}%")
print(f"TA+SA Accuracy: {sa_ta_acc*100:.1f}%")

#features and their coefficients 
print("\nFeature Coefficients")
for feat, coef in sorted(zip(feature_cols_all, model_all.coef_[0]), key=lambda x: abs(x[1]), reverse=True):
    print(f"  {feat:<22} {coef:+.4f}")

#clasification report
print("\n TA + SA classification report:")
print(classification_report(y_test_all, preds_all, zero_division=0))


#confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test_all, preds_all))

#----plot settings-----
close_test_all = data_set["Close"].iloc[-len(y_test_all):].values
dates_test_all = data_set["Date"].iloc[-len(y_test_all):].values
y_test_vals = y_test_all.values

plt.figure(figsize=(14, 6))
plt.plot(dates_test_all, close_test_all, label="Close Price", color="black", linewidth=2)

for i, pred in enumerate(preds_all):
    if i + 1 >= len(dates_test_all):
        break
    color = "green" if pred == 1 else "red"
    marker = '^' if pred == 1 else 'v'
    alpha = 1.0 if pred == y_test_vals[i] else 0.35
    plt.scatter(dates_test_all[i+1], close_test_all[i+1], color=color, s=150, marker=marker, alpha=alpha)

plt.title("TA + SA predictions")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45, ha='right', fontsize=8)

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0],[0], color='black', linewidth=2, label='Close Price'),
    plt.scatter([],[],color='green', marker='^', s=100, label='Predicted Up'),
    plt.scatter([],[],color='red', marker='v', s=100, label='Predicted Down'),
]
plt.legend(handles=legend_elements)
plt.grid(True)
plt.tight_layout()
plt.show()
```
